```python
import string
df['tagline'] = df['tagline'].str.translate(
str.maketrans('', '', string.punctuation))
```

Handling text data (removing stop words)

```python
df.head()
```

 Stop words are common words like very, some,
between etc. To save up the space and speed up the process of prediction, these stop words needs to be removed using NLTK
corpus.

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
```

```python
df.head()
```

```python
def remove_stopwords(tagline):
  return " ".join([word for word in str(tagline).split() if word not in stop_words])
df['tagline']= df['tagline'].apply(lambda x: remove_stopwords(x))
```

```python
df['tagline'].value_counts()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

lemmatization: is a text pre-processing technique used in natural language processing (NLP) models to break a word down to its root meaning to identify similarities. For example, a lemmatization algorithm would reduce the word better to its root word, or lemme, good.


```python
lemmatizer = WordNetLemmatizer()
def lemmatize_words(tagline):
  return" ".join([lemmatizer.lemmatize(word) for word in tagline.split()])
df['tagline'] = df['tagline'].apply(lambda tagline: lemmatize_words(tagline))
```
